```python
categorical_features = data.select_dtypes(include="object")
categorical_features.head(10)
```

```python

def cat_feat_unique_values(categorical_features):
    for feature in categorical_features.columns:
        unique_values_counts = categorical_features[feature].value_counts()
        num_unique_values = unique_values_counts.count()
        
        print(f"Feature: '{feature}', Unique values: {num_unique_values}")
        print("Count of unique values:")
        print(unique_values_counts)
        print()  
categorical_feature = data.select_dtypes(include="object")
cat_feat_unique_values(categorical_features=categorical_feature)
```

```python
binary_arrear_indicator = {"N" : 0 , "Y" : 1}
gender_indicator = {"M" : 0 , "F" : 1, "U" : 2}
chq_acc_ind_indicator = {"N" : 0 ,"Y" : 1}
title_indicator = {"MR": 0 , "MRS" : 1 , "UNKNOWN" : 2, "MS": 3, "AC" : 4 , "MASTER" : 5, "REL" : 6,"MISS": 7 }

indicator_dict = {
    "arrear_ind": binary_arrear_indicator,
    "chq_acc_ind": chq_acc_ind_indicator,
    "gender": gender_indicator,
    "title": title_indicator,
}
categorical_features = ["arrear_ind","chq_acc_ind","gender","title"]
# Замена значений в DataFrame
for col in categorical_features:
    if col in data.columns:  # Проверяем, существует ли столбец
        data[col] = data[col].map(indicator_dict[col])
data.head()
```

```python
data.info()
```

```python
data["Change_In_ATM_TR"].fillna(data["Change_In_ATM_TR"].mean(),inplace=True)
```

```python
new_file_path = '../../data/abt_em_transforme_updated.csv'
data.to_csv(new_file_path, index=False)
```
