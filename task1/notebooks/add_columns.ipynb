```python
import pandas as pd
```

```python

file_path = "/home/xlordplay/alfa_test_tasks/alfa_test/data/Input data/data_prep.tab"  

df = pd.read_csv(file_path, sep="\t")

df.head()
```

```python
# Приведение SEGM_DATE к datetime
#df["SEGM_DATE"] = pd.to_datetime(df["SEGM_DATE"], format="%d%b%Y:%H:%M:%S", errors="coerce")

# Сортировка данных по клиенту и дате
df.sort_values(by=["CLIENT_ID", "SEGM_DATE"], inplace=True)

# Определяем числовые колонки (кроме CLIENT_ID и SEGM_DATE)
num_columns = [col for col in df.columns if col not in ["CLIENT_ID", "SEGM_DATE"]]

# Создаём копию списка колонок для порядка
original_columns = df.columns.tolist()
new_columns = []

for col in original_columns:
    new_columns.append(col)
    if col in num_columns:
        df[f"{col}_1"] = df.groupby("CLIENT_ID")[col].shift(1)  # Предыдущий месяц
        df[f"{col}_2"] = df.groupby("CLIENT_ID")[col].shift(2)  # Предпредыдущий месяц
        new_columns.extend([f"{col}_1", f"{col}_2"])

# Переупорядочиваем DataFrame
df = df[new_columns]



# Сохранение результата
output_file = "../data/data_prep_transforme.csv"  # Укажи путь для сохранения
df.to_csv(output_file, index=False)

print(f"Файл сохранён: {output_file}")
```
